```python
import pandas as pd

SaberPro = pd.read_csv("../SaberPro_limpio.csv")
```

```python
df_SaberPro = SaberPro
print(f"{SaberPro.shape[0]/1000000} millones de filas")
```

```python
SaberPro.columns
```

```python
df_SaberPro = df_SaberPro[['ESTU_GENERO', 'FAMI_EDUCACIONPADRE', 'FAMI_TIENEAUTOMOVIL', 
                           'FAMI_TIENELAVADORA', 'FAMI_ESTRATOVIVIENDA', 'FAMI_TIENECOMPUTADOR', 
                           'FAMI_TIENEINTERNET', 'FAMI_EDUCACIONMADRE', 
                           'MOD_RAZONA_CUANTITAT_PUNT', 'MOD_COMUNI_ESCRITA_PUNT', 
                           'MOD_LECTURA_CRITICA_PUNT', 'MOD_INGLES_PUNT', 
                           'MOD_COMPETEN_CIUDADA_PUNT', 'EDAD_EXAMEN']]
```

```python
df_SaberPro = df_SaberPro.copy()

columnas_puntaje = [
    'MOD_RAZONA_CUANTITAT_PUNT', 
    'MOD_COMUNI_ESCRITA_PUNT', 
    'MOD_LECTURA_CRITICA_PUNT', 
    'MOD_INGLES_PUNT', 
    'MOD_COMPETEN_CIUDADA_PUNT'
]

df_SaberPro['PUNT_GLOBAL'] = df_SaberPro[columnas_puntaje].sum(axis=1)
df_SaberPro = df_SaberPro.drop(columns=columnas_puntaje)
df_SaberPro.head()
```

```python
# Imprimir value_counts() de cada columna en df_SaberPro
for column in df_SaberPro.columns:
    print(f"Value counts para la columna '{column}':")
    print(df_SaberPro[column].value_counts())
    print("\n" + "="*50 + "\n")  # Separador para mayor claridad entre columnas
```

```python
df_SaberPro = df_SaberPro[~df_SaberPro['FAMI_EDUCACIONPADRE'].isin(["No sabe", "No Aplica"])]
df_SaberPro = df_SaberPro[~df_SaberPro['FAMI_EDUCACIONMADRE'].isin(["No sabe", "No Aplica"])]

educacion_ordenada = {
    "Ninguno": 0,
    "Primaria incompleta": 1,
    "Primaria completa": 2,
    "Secundaria (Bachillerato) incompleta": 3,
    "Secundaria (Bachillerato) completa": 4,
    "Técnica o tecnológica incompleta": 5,
    "Técnica o tecnológica completa": 6,
    "Educación profesional incompleta": 7,
    "Educación profesional completa": 8,
    "Postgrado": 9
}

df_SaberPro['FAMI_EDUCACIONPADRE'] = df_SaberPro['FAMI_EDUCACIONPADRE'].map(educacion_ordenada)
df_SaberPro['FAMI_EDUCACIONMADRE'] = df_SaberPro['FAMI_EDUCACIONMADRE'].map(educacion_ordenada)
```

```python
df_SaberPro = df_SaberPro[df_SaberPro['FAMI_ESTRATOVIVIENDA'] != "Sin Estrato"]

estrato_ordenado = {
    "Estrato 1": 1,
    "Estrato 2": 2,
    "Estrato 3": 3,
    "Estrato 4": 4,
    "Estrato 5": 5,
    "Estrato 6": 6
}

df_SaberPro['FAMI_ESTRATOVIVIENDA'] = df_SaberPro['FAMI_ESTRATOVIVIENDA'].map(estrato_ordenado)
```

```python
df_SaberPro['ESTU_GENERO'] = df_SaberPro['ESTU_GENERO'].map({'F': 0, 'M': 1})
df_SaberPro['FAMI_TIENEAUTOMOVIL'] = df_SaberPro['FAMI_TIENEAUTOMOVIL'].map({'No': 0, 'Si': 1})
df_SaberPro['FAMI_TIENELAVADORA'] = df_SaberPro['FAMI_TIENELAVADORA'].map({'No': 0, 'Si': 1})
df_SaberPro['FAMI_TIENECOMPUTADOR'] = df_SaberPro['FAMI_TIENECOMPUTADOR'].map({'No': 0, 'Si': 1})
df_SaberPro['FAMI_TIENEINTERNET'] = df_SaberPro['FAMI_TIENEINTERNET'].map({'No': 0, 'Si': 1})
```

```python
# Imprimir value_counts() de cada columna en df_SaberPro
for column in df_SaberPro.columns:
    print(f"Value counts para la columna '{column}':")
    print(df_SaberPro[column].value_counts())
    print("\n" + "="*50 + "\n")  # Separador para mayor claridad entre columnas
```

### Tipos de Variables Seleccionadas en `df_SaberPro`

A continuación, se clasifica cada variable en `df_SaberPro` según su tipo (binaria, categórica ordinal o nominal, o continua). La variable objetivo es `PUNT_GLOBAL`.

- **ESTU_GENERO**: Binaria. Representa el género del estudiante, con valores **0** (femenino) y **1** (masculino).
- **FAMI_EDUCACIONPADRE**: Categórica ordinal. Nivel educativo del padre, con un orden jerárquico que va desde **Ninguno** (0) hasta **Postgrado** (valor más alto).
- **FAMI_TIENEAUTOMOVIL**: Binaria. Indica si hay automóvil en el hogar, con valores **0** (No) y **1** (Sí).
- **FAMI_TIENELAVADORA**: Binaria. Indica si hay lavadora en el hogar, con valores **0** (No) y **1** (Sí).
- **FAMI_ESTRATOVIVIENDA**: Categórica ordinal. Estrato socioeconómico del hogar, ordenado de **1** (más bajo) a **6** (más alto).
- **FAMI_TIENECOMPUTADOR**: Binaria. Indica si hay computador en el hogar, con valores **0** (No) y **1** (Sí).
- **FAMI_TIENEINTERNET**: Binaria. Indica si hay internet en el hogar, con valores **0** (No) y **1** (Sí).
- **FAMI_EDUCACIONMADRE**: Categórica ordinal. Nivel educativo de la madre, con un orden jerárquico similar al del padre, que va desde **Ninguno** (0) hasta **Postgrado** (valor más alto).
- **EDAD_EXAMEN**: Continua. Edad del estudiante al momento del examen, tratada como una variable continua.
- **PUNT_GLOBAL**: Continua. **Variable objetivo**. Representa el puntaje global acumulado en el examen.

Esta clasificación permite identificar el tipo de preprocesamiento y tratamiento que puede aplicarse a cada columna en el modelo.
